```python
ticker_id = 'VND'
DATA_PATH = os.path.abspath('../vn-stock-data/VNX/')
ticker_data = _af.get_pricing_by_path(DATA_PATH + '/' + ticker_id + '.csv', '2021-01-01')
```

```python
ticker_data
```

```python
ticker_data['height'] = ticker_data.apply(lambda x: x['High'] - x['Low'], axis = 1)
ticker_data['body'] = ticker_data.apply(lambda x: abs(x['Close'] - x['Open']), axis = 1)
ticker_data['up_shadow'] = ticker_data.apply(lambda x: (x['High'] - x['Close']) if (x['Close'] > x['Open']) else (x['High'] - x['Open']), axis = 1)
ticker_data['low_shadow'] = ticker_data.apply(lambda x: (x['Open'] - x['Low']) if (x['Close'] > x['Open']) else (x['Close'] - x['Low']), axis = 1)
ticker_data
```
