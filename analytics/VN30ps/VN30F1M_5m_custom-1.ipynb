### Load Price Data

```python
import os
from pathlib import Path
notebook_path = os.getcwd()
algo_dir = Path(notebook_path).parent.parent
csv_file = str(algo_dir) + '/vn-stock-data/VN30ps/VN30F1M_5minutes.csv'
is_file = os.path.isfile(csv_file)
if is_file:
    dataset = pd.read_csv(csv_file, index_col='Date', parse_dates=True)
else:
    dataset = pd.read_csv("https://raw.githubusercontent.com/zuongthaotn/vn-stock-data/main/VN30ps/VN30F1M_5minutes.csv", index_col='Date', parse_dates=True)
```

```python
data = dataset.copy()
```

```python
data
```

```python
data['rate_x'] = data.apply(
        lambda x: (1 if (x["High"] == x["Low"]) else (x["High"] - x["Open"]) / (x["High"] - x["Low"])), axis=1)
data['rate_y'] = data.apply(
        lambda x: (1 if (x["High"] == x["Low"]) else (x["High"] - x["Close"]) / (x["High"] - x["Low"])), axis=1)
data['ibs'] = data.apply(
        lambda x: (1 if (x["High"] == x["Low"]) else (x["Close"] - x["Low"]) / (x["High"] - x["Low"])), axis=1)
```

```python
data
```

```python
data['rate_x_s1'] = data['rate_x'].shift(1)
data['rate_x_s2'] = data['rate_x'].shift(2)
data['rate_y_s1'] = data['rate_y'].shift(1)
data['rate_y_s2'] = data['rate_y'].shift(2)
data['ibs_s1'] = data['rate_x'].shift(1)
data['ibs_s2'] = data['rate_x'].shift(2)
data
```
